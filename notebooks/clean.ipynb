### Подсчитываем количество nan  в строке и записываем результат в отдельный столюец, если в строке значения наименование только одно, то количество nan будет = количество слов - 1

```python
index_df['nan'] = index_df.isnull().sum(axis = 1)
index_df
```

### Удаляем все строки, где наименования не повторяются, удаляем контрольный столбец nan

```python
clean_index = index_df.loc[index_df['nan'] < words_count]
clean_index.pop('nan')



#books_df = clean_index.index.isin(books_idx)
#target_df = clean_index[~books_df]
#target_df = target_df.drop('nan', axis=1)
#target_df
clean_index
```

# Объединяем пересекающиеся строки в одну строку

```python
levels = clean_index.index
for num, level in enumerate(levels):
    for sublevel in levels[num:]:
        if level == sublevel:
            continue
        print(f'Compare level = {level} & sublevel = {sublevel}')
        diff_idx = clean_index.loc[level].dropna().isin(clean_index.loc[sublevel].dropna())
        if diff_idx.any():
            print(f'row {level} == row {sublevel}')
            tmp_df = pd.DataFrame([clean_index.loc[sublevel], clean_index.loc[level]]).T
            tmp_df['nan'] = np.nan
            tmp_df = tmp_df.fillna(method='ffill',axis=1).T
            clean_index.loc[sublevel] = tmp_df.loc['nan']
            clean_index.drop([level], inplace=True)
            print('drop')
            break
        else:
            print(f'row {level} != row {sublevel}')
print('helllo')        
clean_index
```

## Организуем поиск одинаковых окофов в кластере, формируем индексы кластеров с одинаковыми окофами

### Создается маска значений отличных от nan

```python
mask = xl.notnull()
```

### Накладываем маску похожих названий на их окофы


```python
real_okof = okof[mask]
```

## Организуем поиск одинаковых окофов в кластере, формируем индексы кластеров с одинаковыми окофами

```python
dubl = []
for i in range(3701):
    #print(set(real_okof.loc[i].dropna().to_list()))
    if len(set(real_okof.loc[i].dropna().to_list())) == 1:
        dubl.append(i)
```

## Создаем датафрейм окоф где номера не дублируются

```python
dif_okof = okof.query(f"index not in {dubl}")
```

## Извлекаем индексы (те что пересекаются) в оторых есть отличия по контрольным значениям (окоф, срок службы и т.д.)
