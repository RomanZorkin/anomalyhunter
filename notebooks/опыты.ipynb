```python
import difflib

import numpy as np
import pandas as pd
```

## Задаем список основных средств

```python
words = {
    'Строки':[
        'IP видеокамера B55-5Н',
        'IP видеокамера B55-5Н',
        'IP видеодомофон',
        'Абонентская сеть',
        'Автомобиль "КАМАЗ" в/т4526 №78998654132',
        'USB Накопитель Flash Transcend 8Gb JetFlash 300',
        'Ящик для песка сборно-разборный (0,5 куб. м.)',
        'Ящик для песка сборно-разборный (0,5 куб. м.)',
        'Ящик-носилки для мяса и рыбы',
        'Ящик-носилки для мяса и рыбы',
        'Носилки для больных',
        'Носилки для раненных',
    ],
    'ОКОФ':[
        '2654489494',
        '2654489494',
        '2654489494',
        '2654489494',
        '2654489494',
        '2654489494',
        '2654489494',
        '2654489494',
        '2654489494',
        '2654489494',
        '2654489494',
        '2654489494',
    ],
}

df = pd.DataFrame(words)
df
```

## Преобразование строки в список - далее извлечем из списка заданное количество слов

```python
df['Список'] = df['Строки'].str.split()
df
```

## Преобразуем список в строку с заданным количесвтом слов

```python
df['Short'] = ''
# длина строки 6 слов
depth = 6
for x in range(df.shape[0]):
    words_arr = list(df.iloc[x]['Список'])
    
    if len(words_arr) > depth:
        words_arr = words_arr[:depth]
    df.loc[x, 'Short'] = ' '.join(words_arr)
df
```

## Пустой dataframe, размер = кол-во слов * кол-во слов

```python
idx = df['Short'].to_list()
idx = range(len(idx))
new_df = pd.DataFrame(columns=idx, index=idx)
new_df
```

## Заполняем матрицу (dataframe) весами вероятности совпадения названий

```python
size  = df.shape[0]

for x in range(size):
    for y in range(size):
        row = df.iloc[x]['Short']
        col = df.iloc[y]['Short']
        diff = difflib.SequenceMatcher(None, row, col).ratio()
        new_df.loc[x, y] = diff

new_df[new_df[:] > 0.4]
```

## координаты матрицы (dataframe), удовлетворяющие уровню вероятности совпадения названий, заполняется соответствующими индексами названий

```python
index_df = pd.DataFrame(np.where(new_df > 0.4, new_df.columns, np.nan))
index_df
```

## Объединяем пересекающиеся строки в одну строку



```python
levels = clean_index.index
for num, level in enumerate(levels):
    for sublevel in levels[num:]:
        if level == sublevel:
            continue
        print(f'Compare level = {level} & sublevel = {sublevel}')
        diff_idx = clean_index.loc[level].dropna().isin(clean_index.loc[sublevel].dropna())
        if diff_idx.any():
            print(f'row {level} == row {sublevel}')
            tmp_df = pd.DataFrame([clean_index.loc[sublevel], clean_index.loc[level]]).T
            tmp_df['nan'] = np.nan
            tmp_df = tmp_df.fillna(method='ffill',axis=1).T
            clean_index.loc[sublevel] = tmp_df.loc['nan']
            clean_index.drop([level], inplace=True)
            print('drop')
            break
        else:
            print(f'row {level} != row {sublevel}')
print('helllo')        
clean_index
```
